```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AirQuality.csv' , sep=';')
df1 = pd.read_csv('Heart.csv.csv')
```

# Data Cleaning

```python
df.head()
```

```python
df1.head()
```

```python
df.drop(columns = ['Unnamed: 15','Unnamed: 16'] , inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.replace(to_replace = ',', value='.', regex= True, inplace=True)
df
```

```python
df['Date']= pd.to_datetime(df['Date'])
df['CO(GT)'] = pd.to_numeric(df['CO(GT)'])
df['PT08.S1(CO)'] =pd.to_numeric(df['PT08.S1(CO)'])
df['NMHC(GT)'] = pd.to_numeric(df['NMHC(GT)'])
df['C6H6(GT)'] = pd.to_numeric(df['C6H6(GT)'])
df['PT08.S2(NMHC)'] =pd.to_numeric(df['PT08.S2(NMHC)'])
df['NOx(GT)'] = pd.to_numeric(df['NOx(GT)'])
df['PT08.S3(NOx)'] = pd.to_numeric(df['PT08.S3(NOx)'])
df['NO2(GT)'] = pd.to_numeric(df['NO2(GT)'])
df['PT08.S4(NO2)'] = pd.to_numeric(df['PT08.S4(NO2)'])
df['PT08.S5(O3)'] = pd.to_numeric(df['PT08.S5(O3)'])
df['T'] = pd.to_numeric(df['T'])
df['RH'] =pd.to_numeric(df['RH'])
df['AH'] = pd.to_numeric(df['AH'])
```

```python
df.dtypes
```

```python
mean_1 = df['Date'].mean()
mean_3 = df['CO(GT)'].mean()
mean_4 = df['PT08.S1(CO)'].mean()
mean_5 = df['NMHC(GT)'].mean()
mean_6 = df['C6H6(GT)'].mean()
mean_7 = df['PT08.S2(NMHC)'].mean()
mean_8 = df['NOx(GT)'].mean()
mean_9 = df['PT08.S3(NOx)'].mean()
mean_10 = df['NO2(GT)'].mean()
mean_11 = df['PT08.S4(NO2)'].mean()
mean_12 = df['PT08.S5(O3)'].mean()
mean_13 = df['T'].mean()
mean_14 = df['RH'].mean()
mean_15 = df['AH'].mean()
```

```python
df['Date'].fillna(mean_1,inplace=True)
df['CO(GT)'].fillna(mean_3,inplace=True)
df['PT08.S1(CO)'].fillna(mean_4,inplace=True)
df['NMHC(GT)'].fillna(mean_5,inplace=True)
df['C6H6(GT)'].fillna(mean_6,inplace=True)
df['PT08.S2(NMHC)'].fillna(mean_7,inplace=True)
df['NOx(GT)'].fillna(mean_8,inplace=True)
df['PT08.S3(NOx)'].fillna(mean_9,inplace=True)
df['NO2(GT)'].fillna(mean_10,inplace=True)
df['PT08.S4(NO2)'].fillna(mean_11,inplace=True)
df['PT08.S5(O3)'].fillna(mean_12,inplace=True)
df['T'].fillna(mean_13,inplace=True)
df['RH'].fillna(mean_14,inplace=True)
df['AH'].fillna(mean_15,inplace=True)
```

```python
df.drop(columns=['Time'],inplace=True)
```
